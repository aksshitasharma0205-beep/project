```python

#bar chart
import matplotlib.pyplot as plt

categories = ['Apples', 'Bananas', 'Cherries', 'Dates', 'Elderberries']
values = [10, 15, 7, 12, 5]


plt.bar(categories, values, color='skyblue')

plt.title('Fruit Quantity')
plt.xlabel('Fruit')
plt.ylabel('Quantity')


plt.show()
```

```python
#pie chart
import matplotlib.pyplot as plt
labels = ['Python', 'Java', 'C++', 'Ruby']
sizes = [45, 30, 15, 10]  # Percentages
colors = ['blue', 'yellowgreen', 'red', 'lightskyblue']
explode = (0.1, 0, 0, 0)  

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)

plt.title(" popular python")
plt.axis('equal')  
plt.show()
```

```python
#multipleline graph
import matplotlib.pyplot as plt
days = [1, 2, 3, 4, 5]
score_A = [10, 15, 20, 25, 30]
score_B = [5, 10, 15, 20, 25]

plt.plot(days, score_A, label='Player A')
plt.plot(days, score_B, label='Player B')


plt.title('Scores Over Days')
plt.xlabel('Day')
plt.ylabel('Score')
plt.legend()  
plt.show()
```

```python
#subplot(multiple graph)


import numpy as np

x = np.linspace(0, 10, 100)
y1 = x
y2 = x**2
y3 = np.sin(x)
y4 = np.cos(x)

plt.figure(figsize=(10,8))  # optional: size of the figure

# First subplot
plt.subplot(2,2,1)
plt.plot(x, y1, 'r')
plt.title("Linear")

# Second subplot
plt.subplot(2,2,2)
plt.plot(x, y2, 'g')
plt.title("Quadratic")

# Third subplot
plt.subplot(2,2,3)
plt.plot(x, y3, 'b')
plt.title("Sine")

# Fourth subplot
plt.subplot(2,2,4)
plt.plot(x, y4, 'm')
plt.title("Cosine")

plt.tight_layout()  
plt.show()
```

```python
#Area plot


x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]


plt.fill_between(x, y, color="cyan")  
plt.plot(x, y, color="blue")            

plt.title(" Area Plot")
plt.show()
```
